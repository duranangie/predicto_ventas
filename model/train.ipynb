```python
import pandas as pd
import matplotlib.pyplot as plt # hacer graficas
import seaborn as sns # hacer graficas mas bonitas
```

```python
df = pd.read_csv('../data/Walmart_Sales.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')  # -> Convertir a date

print(df.isnull().sum)
print(df.dtypes)
```

```python
# Agrupar ventas por fecha
sales_by_date = df.groupby('Date')['Weekly_Sales'].sum()
```

```python
sns.set_theme(style="whitegrid")  # Le da estilo bonito al grafico
plt.figure(figsize=(14,6)) #define el tamaÃ±o de la imagen
plt.plot(sales_by_date,sales_by_date.values, label="Ventas semanales totales") # Dibuja la linea de ventas con el tiempo
plt.title('ventas totales a lo largo del tiempo') # Poner el titulo del grafico
plt.xlabel('Fecha') #Etique ta de los ejes
plt.ylabel("ventas")
plt.legend() # Muestra la leyenda
plt.grid(True) # Agrega lineas de cuadricula
plt.tight_layout() #Ajustar el grafico para que no se corte
plt.show() # Muestra el grafico

```
